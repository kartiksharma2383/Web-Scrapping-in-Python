```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
#request library is used to get request from the url
page = requests.get(url)
```

```python
# BeautifulSoup used to parse HTML and XML documents and extract data from them.
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# soup.prettify() used to visualize the parse tree of your HTML content by adding indentation to make it more readable
print(soup.prettify())
```

```python
# used to find first div tag
soup.find('div')
```

```python
# used to find all div tags with class "col-md-12" and soup.find_all('th') and soup.find_all('td')
soup.find_all('div', class_ = 'col-md-12')
```

```python
# used to find all paragraph (<p>) tags that have the class attribute set to 'lead'
soup.find_all('p', class_ = 'lead')
```

```python
soup.find('p', class_ = 'lead').text.strip()
```

```python
# used to finds all table header (<th>) tags on the webpage
soup.find_all('th')
```

```python
# used to find all table data (<td>) tags on the webpage
soup.find_all('td')
```

```python
# used to find the first table header (<th>) tag on the webpage using soup.find('th'). Then, .text extracts the text content within that tag, and .strip() removes any leading or trailing whitespace
soup.find('th').text.strip()
```
